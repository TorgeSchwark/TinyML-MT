```python
import os
import requests
import random
import time

ACCESS_KEY = "uFgQ-DjZNLlyvTnaBX6XZH5K7Vhc7-ewoGLt8R_aszs"
SAVE_DIR = "../huggingface/yolo_backgrounds"
os.makedirs(SAVE_DIR, exist_ok=True)

search_terms = [
    "indoor background", "empty shelf", "room corner", "living room",
    "hallway", "blank wall", "backdrop", "office", "store aisle", "empty table"
]

per_page = 30
target_total = 300
downloaded = 0
page = 1

print(f"Downloading {target_total} general background images...")

while downloaded < target_total:
    term = random.choice(search_terms)
    url = f"https://api.unsplash.com/search/photos?query={term}&page={page}&per_page={per_page}&client_id={ACCESS_KEY}"
    response = requests.get(url)
    data = response.json()

    if 'results' not in data or len(data['results']) == 0:
        print(f"No results for '{term}', skipping...")
        page += 1
        continue

    for img in data['results']:
        img_url = img['urls']['regular']
        img_id = img['id']
        img_data = requests.get(img_url).content
        with open(os.path.join(SAVE_DIR, f"{img_id}.jpg"), 'wb') as f:
            f.write(img_data)
        downloaded += 1
        print(f"Downloaded {downloaded}/{target_total}")
        if downloaded >= target_total:
            break
    page += 1
    time.sleep(0.5)  # polite delay to avoid rate limits
```
